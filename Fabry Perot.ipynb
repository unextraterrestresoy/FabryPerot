```python
# Import libraries
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

# Fabry-Perot interferometer

Source> [Link to PDF FabryPerot](https://physics.iitm.ac.in/~ph5060/manuals/FabryPerot)

```python
(4*0.36)/((1-0.36)**2)
```

```python
# delta es la fase
delta = np.linspace(0,2*np.pi,150) 
def f1(x,r):
    'Function to calculate the transmitted intensity I_T'
    'given in terms of the insident wave intensity I_0 = 1 for simplicity'
    # r es la reflectivity de los espejos
    f = (4*r)/((1-r)**2)               # f is the coefficient of finesse
    i1= 1 /(1+f*(np.sin(1.5+x))**2)        # Transmitted intensity
    return i1
```

```python
nombre=[r'$R = 0.1$' , '$R = 0.36$', '$R = 0.85$']
plt.figure(figsize=(11,8))
plt.plot(delta, f1(delta,0.1), "-k.",label=r'%s'%nombre[0])
plt.plot(delta, f1(delta,0.36), "-k.",label=r'%s'%nombre[1])
plt.plot(delta, f1(delta,0.85), "-k.",label=r'%s'%nombre[2])
plt.ylabel("$I_{0}$ [arb. units]",fontsize=16)
plt.xlabel("$\delta$ [$p \pi$]",fontsize=16)
plt.title("Fabry-Perot transmission as a function of $\delta$", fontsize=16)
plt.ylim(0,1.2)
plt.legend(loc='best',fontsize=16)
plt.grid()
plt.show()
```
